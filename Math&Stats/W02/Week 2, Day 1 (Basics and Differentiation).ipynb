### Solving equations (and systems of equations)

```python
# Defining equations with sympy.Eq() and solving them with sympy.solve()

# 2x - 1 = 4
eq_1 = sp.Eq(2*x - 1, 4)
sol_1 = sp.solve(eq_1) #solutions are given as a list
print(sol_1)

# x/3 + 2 = 0
eq_2 = sp.Eq(x/3 + 2, 0)
sol_2 = sp.solve(eq_2)
print(sol_2)

# sin(x) = -cos(2x)
eq_3 = sp.Eq(sp.sin(x), -sp.cos(2*x))
sol_3 = sp.solve(eq_3)
print(sol_3)

# x*ln(x) = 0
eq_4 = sp.Eq(sp.ln(x)**2 - sp.ln(x**2), 0)
sol_4 = sp.solve(eq_4)
print(sol_4)
```

#### Example: solve the following equations
* $x^3 - 2 x^2 + 1 = 0$
* $e^{-x^2} - e^x = 0$

```python
# x^3 - 2*x^2 + 1 = 0
eq_1 = sp.Eq(x**3 - 2*x**2 + 1, 0)
sol_1 = sp.solve(eq_1)
print(sol_1)
print(sol_1[0], sol_1[1].evalf(), sol_1[2].evalf())

# exp(-x**2) - exp(x)
eq_2 = sp.Eq(sp.exp(-x**2) - sp.exp(x), 0)
sol_2 = sp.solve(eq_2)
print(sol_2)
```

#### Solving systems of linear equations
Solve the system:
\begin{equation}\left\{
\begin{array}{rcl}
-5x + 7y &=& 2\\
3x + 5y &=& 8
\end{array}\right.
\end{equation}

```python
# Solving systems of linear equations

x, y = sp.symbols('x y')

sys_1 = sp.Eq(-5*x + 7*y, 2)
sys_2 = sp.Eq(3*x + 5*y, 8)

sol_sys = sp.solve((sys_1, sys_2), (x, y))

print(sol_sys)

print(sol_sys[x])

print(sol_sys[y])
```

#### Practice: solve the following system of linear equations:
\begin{equation}\left\{

x - 2y + 3z &=& 1\\
-x + 3y + 2z &=& -6\\
2x + y + 3z &=& 5
\end{array}\right.
\end{equation}

## Differentiation of functions
The derivative of a function $y = f(x)$, labeled as $y'$ or $f'(x)$ or $\frac{dy}{dx}$, describes (quatifies) the **instantaneous rate of change** of the function. In other words:
* If the derivative is positive, the function is increasing in value. Conversely, if the derivative is negative, then the function is decreasing in value
* The greater the *number* of the derivative, the greater the rate of increasing or decreasing of the function

### Examples of differentiation

#### Example: finding derivatives and plotting them
* Find the derivative of $p(x) = \sin(x^2)$, and then plot the function and its derivative on same graph on the range $(-2, 2)$
* Find the derivative of $q(x) = e^{x}$, and then plot the function and its derivative on same graph on the range $(-2, 1.3)$
* Find the derivative of $r(x) = e^{\cos x}$, and then plot the function and its derivative on same graph on the range $(-2, 4)$

```python
# function p(x)
p = sp.Function('p', real=True)
p = sp.sin(x**2)
dp = sp.Function('dp', real=True)
dp = p.diff()

p_1 = sp.plot(p, (x, -2, 2), show=False, ylabel='y', line_color='b')
p_2 = sp.plot(dp, (x, -2, 2), show=False, line_color='r')
p_1.append(p_2[0])
p_1.show()
```

```python
# function q(x)
q = sp.Function('q', real=True)
q = sp.exp(x)
dq = sp.Function('dq', real=True)
dq = q.diff()

q_1 = sp.plot(q, (x, -2, 1.3), show=False, ylabel='y', line_color='b')
q_2 = sp.plot(dq, (x, -2, 1.3), show=False, line_color='r')
q_1.append(q_2[0])
q_1.show()
```

```python
# function r(x)
r = sp.Function('r', real=True)
r = sp.exp(sp.cos(x))
dr = sp.Function('dr', real=True)
dr = r.diff()

r_1 = sp.plot(r, (x, -2, 4), show=False, ylabel='y', line_color='b')
r_2 = sp.plot(dr, (x, -2, 4), show=False, line_color='r')
r_1.append(r_2[0])
r_1.show()
```
