## Day 2: Finding minima and maxima of functions
* Differential Calculus is an important tool in **optimization** -- the process of finding **extremal points** (minima and maxima) of functions
* If the derivative is zero at some point, then *most often* the function will have either a local minimum or a local maximum at that point
* To find a minimum or a maximum point of a function $f(x)$, we need to calculate the derivative, and set it equal to 0. The solution(s) to the equation $f'(x) = 0$ are *possible* minimum or maximum points. These are called **critical points**

### Example 1
* Let $f(x) = x^3 - 3x$. Define this function in SymPy and calculate its derivative $f'(x)$
* Solve the equation $f'(x) = 0$ to identify possible extremal points
* By graphing the function, state the minimum and/or maximum points of the function, and calculate the minimum and/or maximum values.

```python
# define variable x
x = sp.symbols('x')

# define function f
f = sp.Function('f', real=True)
f = x**3 - 3*x

# differentiate f
df = sp.Function('df', real=True)
df = f.diff()

# solve f' = 0
ex_pts_f = sp.solve(df, 0, x)
print(f'Possible extremal points are x = {ex_pts_f[0][x]} and x = {ex_pts_f[1][x]}')

# graph the function f
sp.plot(f, (x, -2.5, 2.5), ylabel='y')

# stating the conclusions
print(f'The point x = {ex_pts_f[0][x]} is a local maximum with a maximum value of f({ex_pts_f[0][x]}) = {f.subs(x, ex_pts_f[0][x])}')
print(f'The point x = {ex_pts_f[1][x]} is a local minimum with a minimum value of f({ex_pts_f[1][x]}) = {f.subs(x, ex_pts_f[1][x])}')
```

### Example 2
* Using the same process from Example 1, identify the extremal points of the function $g(x) = x^4 - 2x^3 + x^2 - 1$

```python
# define function g
g = sp.Function('g', real=True)
g = x**4 - 2*x**3 + x**2 - 1

# differentiate g
dg = sp.Function('dg', real=True)
dg = g.diff()

# solve g' = 0
ex_pts_g = sp.solve(dg, 0, x)
print(f'Possible extremal points are x = {ex_pts_g[0][x]}, x = {ex_pts_g[1][x]} and x = {ex_pts_g[2][x]}')

# graph the function g
sp.plot(g, (x, -0.8, 1.8), ylabel='y')

# stating the conclusions
print(f'The point x = {ex_pts_g[0][x]} is a local minimum with a minimum value of g({ex_pts_g[0][x]}) = {g.subs(x, ex_pts_g[0][x])}')
print(f'The point x = {ex_pts_g[1][x]} is a local maximum with a maximum value of g({ex_pts_g[1][x]}) = {g.subs(x, ex_pts_g[1][x])}')
print(f'The point x = {ex_pts_g[2][x]} is a local minimum with a minimum value of g({ex_pts_g[1][x]}) = {g.subs(x, ex_pts_g[2][x])}')
```

### Example 3: "fake" extremal points
* We consider the function $h(x) = x^3$. Let's try to identify extremal points

```python
# define function h
h = sp.Function('h', real=True)
h = x**3

# differentiate h
dh = sp.Function('dh', real=True)
dh = h.diff()

# solve h' = 0
ex_pts_h = sp.solve(dh, 0, x)
print(f'Possible extremal point: x = {ex_pts_h[0][x]}')

# graph the function h
sp.plot(h, (x, -2, 2), ylabel='y')

# print conclusion
print(f'The point x = {ex_pts_h[0][x]} is neither a local minimum, nor a local maximum point :(')
```

### Higher order derivatives
* To avoid relying on the graph when identifying minimum and maximum points, we resort to using *second order derivatives*, or short - using the **second derivative** of the function
* The second derivative of the function $f(x)$, denoted by $f''(x)$ [read: f-double-prime] is just the derivative of the derivative
* The differentiation process could go beyond the second derivative

### Example 4: calculating second derivatives
* Calculate the second derivatives of the functions $f(x)$ and $g(x)$ from the previous examples

```python
# calculating f'' using function.diff().diff()
df2 = sp.Function('df2', real=True)
df2 = f.diff().diff()
df2
```

```python
# calculating g'' using function.diff(variable, 2)
dg2 = sp.Function('dg2', real=True)
dg2 = g.diff('x', 2)
dg2
```

### Identifying extremal points properly: Second Derivative Test
* To identify if an extremal point is a minimum or a maximum, we substitute its value into the second derivative. Let's label an extremal point for the function $f(x)$ by $x^*$. Since this is an extremal point, at that point we have $f'(x^*) = 0$
* If $f''(x^*) > 0$, then $x^*$ is a local minimum
* If $f''(x^*) < 0$, then $x^*$ is a local maximum
* If $f''(x^*) = 0$, then $x^*$ could be either a local maximum or a local minimum, or neither of the two (this is worst case scenario)

### Example 1a: back to $f(x)$
* Using the second derivative test, classify the extremal points of $f(x)$ as minimum or maximum points

```python
# the possible extremal points of f are
print(f'Possible extremal points are x = {ex_pts_f[0][x]} and x = {ex_pts_f[1][x]}')

# evaluate f'' at the possible extremal points
print(f"f''({ex_pts_f[0][x]}) = {df2.subs('x', ex_pts_f[0][x])}. This value is negative, so x = {ex_pts_f[0][x]} is a local maximum")
print(f"f''({ex_pts_f[1][x]}) = {df2.subs('x', ex_pts_f[1][x])}. This value is positive, so x = {ex_pts_f[1][x]} is a local minimum")
```

### Example 2a: back to $g(x)$
* Using the second derivative test, classify the extremal points of $g(x)$ as minimum or maximum points

```python
# the possible extremal points of g are
print(f'Possible extremal points are x = {ex_pts_g[0][x]}, x = {ex_pts_g[1][x]} and x = {ex_pts_g[2][x]}')

# evaluate g'' at the possible extremal points
print(f"g''({ex_pts_g[0][x]}) = {dg2.subs('x', ex_pts_g[0][x])}. This value is positive, so x = {ex_pts_g[0][x]} is a local minimum")
print(f"g''({ex_pts_g[1][x]}) = {dg2.subs('x', ex_pts_g[1][x])}. This value is negative, so x = {ex_pts_g[1][x]} is a local maximum")
print(f"g''({ex_pts_g[2][x]}) = {dg2.subs('x', ex_pts_g[2][x])}. This value is positive, so x = {ex_pts_g[2][x]} is a local minimum")
```

### Example 3a: back to $h(x)$
* Using the second derivative test, classify the extremal points of $h(x)$ as minimum or maximum points

```python
# the possible extremal points of h are
print(f'Possible extremal point is x = {ex_pts_h[0][x]}')

# evaluate h'' at the possible extremal points
print(f"h''({ex_pts_h[0][x]}) = {h.diff('x', 2).subs('x', ex_pts_h[0][x])}. This value is neither positive nor negative, so we cannot say if the point is a local minimum or a local maximum")
```

### Example 5
* Find the extremal points of the function $y = x \cdot e^{x}$

```python
# defining y(x)
y = sp.Function('y', real=True)
y = x*sp.exp(x)

# defining y'(x)
dy = sp.Function('dy', real=True)
dy = y.diff()

# defining y''(x)
dy2 = sp.Function('dy2', real=True)
dy2 = y.diff('x', 2)

# getting the potential extremal points
ex_pts_y = sp.solve(dy, 0, 'x')
print(f'There is only one potential extremal point, x = {ex_pts_y[0][x]}')

# cheching y''(extremal)
print(f"Since y''({ex_pts_y[0][x]}) =  {dy2.subs('x', ex_pts_y[0][x])} = {dy2.subs('x', ex_pts_y[0][x]).round(3)}, we conclude that x = {ex_pts_y[0][x]} is a local minimum")
print(f"The minimum value is y({ex_pts_y[0][x]}) = {y.subs('x', ex_pts_y[0][x])} = {y.subs('x', ex_pts_y[0][x]).round(3)}")
```

### Example 6: using embedded SymPy functions to find minimum/maximum values
* SymPy offers two embedded function functions for finding minimum and maximum **values**. These are sympy.minimum() and sympy.maximum()

```python
# find the minimum of the quadratic function y = x^2 + x + 1
print(f'The minimum value is {sp.minimum(x**2 + x + 1, x)}')

# find the maximum of the function 1 - (ln(x) - 1)^2
print(f'The maximum value is {sp.maximum(1 - (sp.ln(x) - 1)**2, x)}')
```
