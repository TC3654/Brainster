### Exercise 1

Generate your own array of numbers and plot the natural logarithmic and exponential plot for x in range [0, n).  

np.log(some_array) # log every element  
np.exp2(some_array) # 2**elem every element

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def log(n):
    l = [x for x in range(1, n+1)]
    log_n = np.log(l)
    plt.plot(l, log_n)

def exp(n):
    l = [x for x in range(1, n+1)]
    exp_n = np.exp2(l)
    plt.plot(l, exp_n)      
```

```python
log(5)
```

```python
exp(10)
```

```python
#Solution from class (Kuba)
ar = np.log(range(1,320))
plt.plot(ar)
plt.show()
ar2 = np.exp2(range(1,10))
plt.plot(ar2)
plt.show()
```

### Exercise 2

- On the Amazon fires dataset, create a bar plot which shows the total fires per each state.  
- On the Amazon fires dataset, create a graph plot to show the fires in Sao Paulo state per year

```python
fires_df = pd.read_csv("amazon.csv")
fires_df.head()
```

```python
by_state = fires_df.groupby("state")
fires_by_state = by_state["number_of_fires"].sum()
fires_by_state.plot(kind="bar")
```

```python
grp = fires_df.groupby(["state", "year"])["number_of_fires"].sum()
sao_paulo = grp["Sao Paulo"]
plt.plot(sao_paulo.index, sao_paulo.values)
```
